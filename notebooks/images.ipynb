```python
WIDTH = 1280
HEIGHT = 720
maze = cv.imread('../images/maze/m%d.png' % (1), cv.IMREAD_GRAYSCALE)
maze_array = cv.resize(maze, (WIDTH, HEIGHT), interpolation=cv.INTER_NEAREST)
inverted_maze_array = cv.bitwise_not(maze_array)
dust = np.ones((HEIGHT, WIDTH))
np.sum(dust)
```

```python
idx = np.where(inverted_maze_array > 0)
dust[idx] = 0
np.sum(dust)
```
