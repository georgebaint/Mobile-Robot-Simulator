```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from settings import *
```

```python
image = cv.imread("images/maze.png", cv.IMREAD_GRAYSCALE)
# Apply a binary threshold to distinguish walls from free space
_, binary_image = cv.threshold(image, 127, 255, cv.THRESH_BINARY_INV)
# Convert binary image to match the display dimensions
maze = cv.resize(binary_image, (WIDTH, HEIGHT))
```

```python
maze.shape
```

```python
plt.imshow(binary_image, cmap='gray')
```

```python
circle = np.zeros((720, 1280), dtype='uint8')
circle = cv.circle(circle, (100, 100), 60, 255, -1)
plt.imshow(circle, cmap='gray')
```

```python
res = circle ^ maze
plt.imshow(res, cmap='gray')
```

```python
inter = circle & maze
coords = np.argwhere(inter > 0)
coords
```

```python
ys = coords[:, 0]
xs = coords[:, 1]
range = (np.max(xs) - np.min(xs), np.max(ys) - np.min(ys))
range
```
