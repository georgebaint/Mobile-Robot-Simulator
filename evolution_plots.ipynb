```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
```

```python
# for maze_num in range(4):
with open('training_generations/Map_%d/genotype_history.pkl' % (2), 'rb') as f:
    genotype_history = pickle.load(f)

with open('training_generations/Map_%d/fitness_history.pkl' % (2), 'rb') as f:
    fitness_history = pickle.load(f)
```

```python
genotype_matrix = np.vstack(genotype_history)

plt.figure(figsize=(15, 10))
plt.imshow(genotype_matrix, aspect='auto', interpolation='none', cmap='viridis')
plt.colorbar(label='Genotype Value')
plt.xlabel('Parameter Index')
plt.ylabel('Generation and Subject Index')
plt.title('Heatmap of Genotype Values Across Generations')
plt.show()
```

```python
average_genotypes = [np.mean(np.array(gen), axis=0) for gen in genotype_history]

average_genotype_matrix = np.vstack(average_genotypes)

plt.figure(figsize=(15, 10))
plt.imshow(average_genotype_matrix, aspect='auto', interpolation='none', cmap='viridis')
plt.colorbar(label='Average Genotype Value')
plt.xlabel('Parameter Index')
plt.ylabel('Generation Index')
plt.title('Heatmap of Average Genotype Values Across Generations')
plt.show()
```

```python
best_fitness = [max(gen) for gen in fitness_history]
average_fitness = [sum(gen) / len(gen) for gen in fitness_history]
worst_fitness = [min(gen) for gen in fitness_history]

generations = list(range(1, len(fitness_history) + 1))

plt.figure(figsize=(10, 5))
plt.plot(generations, best_fitness, label='Best Fitness', marker='o')
plt.plot(generations, average_fitness, label='Average Fitness', marker='o')
plt.plot(generations, worst_fitness, label='Worst Fitness', marker='o')
plt.xlabel('Generation')
plt.ylabel('Fitness Score')
plt.title('Fitness Trends Across Generations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(fitness_history, labels=[f'Generation {i+1}' for i in range(len(fitness_history))])
plt.xlabel('Generation')
plt.ylabel('Fitness Score')
plt.title('Distribution of Fitness Scores per Generation')
plt.show()
```

```python
fitness_matrix = np.array(fitness_history)

plt.figure(figsize=(10, 5))
sns.heatmap(fitness_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.xlabel('Individual')
plt.ylabel('Generation')
plt.title('Heatmap of Fitness Scores Across Generations')
plt.show()
```
